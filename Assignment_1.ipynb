```python
#data cleaning
#by the above head() inspection one can very well see that there are two levels on four wheel drive.
#fwd and 4wd. So here i am converting the 4wd to fwd for uniformity.



df['drivewheel'] = np.where(df['drivewheel']=='4wd', 'fwd', df['drivewheel'])
```

```python
#checking whether the changes made is reflected or not.
df
```

```python
#since CarName doesnt make sense in predicting the price of the cars. 
#here i am splitting the CarCompany name and the model name into two coloumns

new = df["CarName"].str.split(" ", n = 1, expand = True) 
df["CarCompany"]= new[0] 
df["Model"]=new[1]
df
```

```python
#here i am dropping CarName,car_ID and Model. Since they dont make any sense 
#while in the analysis part


df.drop(columns =["CarName"], inplace = True) 
df.drop(columns =["car_ID"], inplace = True) 
df.drop(columns=["Model"],inplace=True)
```

```python
#here i am checking whether the changes made are been reflected.
df
```

```python
#since i doubt that there can be some descrepencies in the carname,
#here i am just cheking all the carnames
#one can clearly see that there are spelling mistakes in some of the entries
#like toyouta,Nissan,vokswagen,vw,porcshce,maxda etc.,

idx = pd.Index(df['CarCompany'])
pd.unique(idx)
```

```python
#here i am correcting those mistakes

df['CarCompany'] = np.where(df['CarCompany']=='maxda', 'mazda', df['CarCompany'])
df['CarCompany'] = np.where(df['CarCompany']=='Nissan', 'nissan', df['CarCompany'])
df['CarCompany'] = np.where(df['CarCompany']=='porcshce', 'porsche', df['CarCompany'])
df['CarCompany'] = np.where(df['CarCompany']=='toyouta', 'toyota', df['CarCompany'])
df['CarCompany'] = np.where(df['CarCompany']=='vw', 'volkswagen', df['CarCompany'])
df['CarCompany'] = np.where(df['CarCompany']=='vokswagen', 'volkswagen', df['CarCompany'])
```

```python
df
```

```python
#upon inspecting the coloumn symboling one can clearly see that it ranges from
#-3 and +3. So here i am convering them to the scale of 0 to 6 . 
#this is being done so as to assist the numerical regression modelling

df['symboling'] = np.where(df['symboling']==3, 6, df['symboling'])
df['symboling'] = np.where(df['symboling']==2, 5, df['symboling'])
df['symboling'] = np.where(df['symboling']==1, 4, df['symboling'])
df['symboling'] = np.where(df['symboling']==0, 3, df['symboling'])
df['symboling'] = np.where(df['symboling']==-1, 2, df['symboling'])
df['symboling'] = np.where(df['symboling']==-2, 1, df['symboling'])
df['symboling'] = np.where(df['symboling']==-3, 0, df['symboling'])  
```

```python
df
```

```python
#CONVERTING ALL THE CATEGORICAL variables into BINARY NUMERIC (0 or 1) variables

df['fueltype'] = np.where(df['fueltype']=='gas', 0,  1)
```

```python
df['aspiration'] = np.where(df['aspiration']=='std', 0, 1)
```

```python
df['doornumber'] = np.where(df['doornumber']=='two', 0, 1)
```

```python
df['enginelocation'] = np.where(df['enginelocation']=='rear', 0, 1)
```

```python
#here i am checking the data types

df.info()
```

```python
#in the next 6 commands i am developing dummy coloumns for each of the CATEGORICAL variables.

status = pd.get_dummies(df['CarCompany'], drop_first = True)
df = pd.concat([df, status], axis = 1)
df.drop(['CarCompany'], axis = 1, inplace = True)
```

```python
status = pd.get_dummies(df['carbody'], drop_first = True)
df = pd.concat([df, status], axis = 1)
df.drop(['carbody'], axis = 1, inplace = True)
```

```python
status = pd.get_dummies(df['drivewheel'], drop_first = True)
df = pd.concat([df, status], axis = 1)
df.drop(['drivewheel'], axis = 1, inplace = True)
```

```python
status = pd.get_dummies(df['enginetype'], drop_first = True)
df = pd.concat([df, status], axis = 1)
df.drop(['enginetype'], axis = 1, inplace = True)
```

```python
status = pd.get_dummies(df['cylindernumber'], drop_first = True)
df = pd.concat([df, status], axis = 1)
df.drop(['cylindernumber'], axis = 1, inplace = True)
```

```python
status = pd.get_dummies(df['fuelsystem'], drop_first = True)
df = pd.concat([df, status], axis = 1)
df.drop(['fuelsystem'], axis = 1, inplace = True)
```

```python
#finally checking all the variable types for one last time.

#everything is now numerical. ALL IS WELL.
df.info()
```
